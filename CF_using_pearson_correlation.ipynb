# Movie Recommendation System

```python
# importing all dependencies
import pandas as pd
```

```python
ratings=pd.read_csv('ratings.csv')
movies=pd.read_csv('movies_metadata.csv',low_memory=False)
```

```python

ratings.drop(columns=['timestamp'], inplace=True)
ratings.head()
```

```python
movies.head()
```

```python
movies.info()
```

```python
selected_columns=['id','title']
movies=movies[selected_columns]
```

```python
movies.head()
```

```python
movies = movies.rename(columns={'id': 'movieId'}).copy()
movies.head()
```

```python
movies.info()
```

```python
ratings.info()
```

```python
# Convert 'movieId' column to numeric, coerce errors to NaN
movies['movieId'] = pd.to_numeric(movies['movieId'], errors='coerce')

# Drop rows where 'movieId' is NaN
movies.dropna(subset=['movieId'], inplace=True)
```

```python
movies['movieId'] = movies['movieId'].astype(int)
merged_data=pd.merge(ratings,movies,on='movieId',how='inner')
merged_data.head()
```

```python
merged_data.shape
```

```python
# Assuming your DataFrame is named 'df'
movie_data = merged_data.sample(n=20000, random_state=4)
movie_data.shape
```

```python
user_ratings= movie_data.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
# let's drop movies which have less than 10 users who rated it. and fill remaining null values with 0
user_ratings =user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings.head(20)
```

```python
# similarity Matrix
item_similarity_df=user_ratings.corr(method='pearson')
item_similarity_df.head(20)
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
print(item_similarity_df.index)
```

```python
# Our Movie Recommendation System
def get_similar_movies(movie_name, user_rating, item_similarity_df):
    similar_score = item_similarity_df[movie_name] * (user_rating - 2.5)
    similar_score = similar_score.sort_values(ascending=False)
    similar_movies_df = pd.DataFrame({'Movie Name': similar_score.index, 'Similarity Score': similar_score.values})
    return similar_movies_df

# Initialize an empty DataFrame to store similar movies
similar_movies = pd.DataFrame()

# Let's make some movie recommendations to a user with some feedback given by the them
movie_list=[('A Bridge Too Far',2),
            ('Yankee Doodle Dandy',1),
            ('A Perfect Murder',4),
            ('A Time to Kill',4),
            ('X-Men Origins: Wolverine',5)]

# Iterate over the list of movie names and user ratings
for movie, rating in movie_list:
    similar_movies = pd.concat([similar_movies, get_similar_movies(movie, rating, item_similarity_df)], ignore_index=True)

# Display the Recommended movies to the user
similar_movies.head(10)
```
